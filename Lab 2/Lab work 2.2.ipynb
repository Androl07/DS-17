```python
import numpy
import math
#6
import random
list_with_random_numbers = [random.randint(0,100) for i in range(1000)]
print("Максимальне число:", max(list_with_random_numbers))
print("Мінімальне число:", min(list_with_random_numbers))

mean = sum(list_with_random_numbers)/len(list_with_random_numbers)
print("Середнє значення послідовності:", mean)

list_with_random_numbers.sort()
median = list_with_random_numbers[len(list_with_random_numbers)//2]
print("Медіана послідовності:", median)

list_with_random_numbers_uniq = []
for i in list_with_random_numbers:
    if i not in list_with_random_numbers_uniq:
        list_with_random_numbers_uniq.append(i)

for i in range(len(list_with_random_numbers_uniq)):
    print("Цифра:", list_with_random_numbers_uniq[i], ", кількість входжень:", list_with_random_numbers.count(i))
    i = i + 1

e = 0
for i in range(len(list_with_random_numbers_uniq)):
    e = e + (i - mean)**2*list_with_random_numbers.count(i)
    i = i + 1
std_my = round(math.sqrt(e/(len(list_with_random_numbers) - 1)), 2)

std = round(numpy.std(list_with_random_numbers, axis = 0), 2)
print("Середнє квадратичне відхилення (власні розрахунки):", std_my, ", Середнє квадратичне відхилення (за формулою):", std)
```
